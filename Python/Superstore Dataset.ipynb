```python
# LEVEL 3 — Advanced (Feature Engineering + Visualization)
# Order processing time = Ship Date – Order Date, new column banao.
Superstore["Order Date"]=pd.to_datetime(Superstore["Order Date"])
Superstore["Ship Date"]=pd.to_datetime(Superstore["Ship Date"])
Superstore["processing_time"]=(Superstore["Ship Date"]-Superstore["Order Date"]).dt.days
Superstore[["Order Date","Ship Date","processing_time"]].head()
```

```python
# Average delivery time for each ship mode calculate karo.
Superstore.groupby("Ship Date")["processing_time"].mean()
```

```python
# Customer lifetime value (CLV) create karo (total sales per customer).
CLV=Superstore.groupby("Customer ID")["Sales"].sum()
CLV.sort_values(ascending=False).head()
```

```python
# RFM segmentation ke liye Recency/ Frequency/ Monetary metrics banao.
# 1st Step-Last purchase date
max_date=Superstore["Order Date"].max()
# 2nd Step-RFM Calculation
RFM=Superstore.groupby("Customer ID").agg({
    "Order Date":lambda x:(max_date-x.max()).days,      #recency
    "Order ID":"nunique",                               #frequency 
    "Sales":"sum"})                                     #monetary
RFM.columns=["Recency","Frequency","Monetary"]
RFM.head()
```

```python
# Pareto Analysis: top 20% customers giving 80% sales identify karo.
# 1st step-Customer-wise total sales
cust_sales=Superstore.groupby("Customer ID")["Sales"].sum().sort_values(ascending=False)
# 2nd step-Cumulative contribution%
pareto=cust_sales.cumsum()/cust_sales.sum()*100
pareto.head()
# 3rd step-Filter top 20% contributors
top_20_percent=pareto[pareto<=80]
top_20_percent
```

```python
# Category-wise monthly trend multi-line chart banao.
# 1st step-month column
Superstore["month"]=Superstore["Order Date"].dt.to_period("m").astype(str)
# 2nd step-groupby category+month
trend=Superstore.groupby(["month","Category"])["Sales"].sum().reset_index()
# 3rd step-multi-line plot
plt.figure(figsize=(8,5))
for cat in trend["Category"].unique():
    subset=trend[trend["Category"]==cat]
    plt.plot(subset["month"],subset["Sales"],label=cat)
plt.xticks(rotation=90)
plt.title("Category-wise Monthly Sales Trend")
plt.xlabel("month")    
plt.ylabel("Sales")    
plt.legend()    
plt.show()    
```

```python
# Sub-category profitability heatmap banao (profit × category matrix).
# 1st step-pivot table
pivot=Superstore.pivot_table(values="Profit",index="Category",columns="Sub-Category",aggfunc="sum")
# 2nd step-heatmap plot
plt.figure(figsize=(8,5))
sns.heatmap(pivot,annot=True,cmap="YlGnBu")
plt.title("Sub-Category Profitability Heatmap")
plt.show()
```

```python
# High-sales–low-profit products detect karne ka code likho.
#criteria:sales>median,profit<median
high_sales=Superstore["Sales"].median()
low_profit=Superstore["Profit"].median()
problem_products=Superstore[(Superstore["Sales"]>high_sales)&(Superstore["Profit"]<low_profit)]
problem_products.head()
```

```python
# State-wise loss contribution percentage calculate karo.
# 1st step-total loss(profit<0)
loss_df=Superstore[Superstore["Profit"]<0]
# 2nd step- state-wise loss sum
state_loss=loss_df.groupby("State")["Profit"].sum().abs()
# 3rd step- contribution%
loss_percentage=(state_loss/state_loss.sum())*100
loss_percentage.sort_values(ascending=False)
```

```python
# LEVEL 4 — Expert / Interview Business Case (GitHub Portfolio Level)
# Predictive analysis ke liye simple sales forecasting (ARIMA/Prophet) model banao.
from statsmodels.tsa.arima.model import ARIMA
#prepare monthly sales series
Superstore["Order Date"]=pd.to_datetime(Superstore["Order Date"])
monthly_sales=Superstore.groupby(Superstore["Order Date"].dt.to_period("m"))["Sales"].sum()
monthly_sales.index=monthly_sales.index.to_timestamp()
# train ARIMA
model=ARIMA(monthly_sales,order=(1,1,1))
model_fit=model.fit()
# forecast next 6 months
forecast=model_fit.forecast(6)
forecast
```

```python
# K-Means clustering: customer segmentation (use Customer ID features).
from sklearn.cluster import KMeans
#RFM from previous step
RFM=Superstore.groupby("Customer ID").agg({
    "Order Date":lambda x:(Superstore["Order Date"].max()-x.max()).days,      
    "Order ID":"nunique",                                
    "Sales":"sum"})                                     
RFM.columns=["Recency","Frequency","Monetary"]
Kmeans=KMeans(n_clusters=5,random_state=0)
RFM["cluster"]=Kmeans.fit_predict(RFM)
RFM.head()
```

```python
# Product recommendation ke liye basic collaborative filtering model banao.
from sklearn.metrics.pairwise import cosine_similarity
#pivot table(customer*products)
matrix=Superstore.pivot_table(index="Customer ID",columns="Product ID",values="Quantity",fill_value=0)
similarity=cosine_similarity(matrix)
similarity_df=pd.DataFrame(similarity,index=matrix.index,columns=matrix.index)
# recommendation for 1 customer eg.
similarity_df.loc["AB-10015"].sort_values(ascending=False).head(10)
```

```python
# Machine Learning pipeline banao for predicting profit category (High/Low).
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
#Create target
Superstore["Profit_Category"]=Superstore["Profit"].apply(lambda x:"High" if x>=Superstore["Profit"].median () else "Low")
# features
x=Superstore[["Sales","Discount","Quantity"]]
y=Superstore["Profit_Category"]
#Pipeline
pipeline=Pipeline([
    ("scaler",StandardScaler())
    ("model",RandomForestClassifier())
])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
pipeline.fit(x_train,y_train)
pipeline.score(x_test,y_test)
```
